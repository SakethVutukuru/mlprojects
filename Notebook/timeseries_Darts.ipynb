```python
import pandas as pd
# Load the CSV file into a pandas Series
series = pd.read_csv('data_daily.csv', squeeze=True)
```

```python
import matplotlib.pyplot as plt


series.plot(label='daily receipts')
plt.xlabel('Day of the year 2021')
plt.ylabel('Receipt_Count')
plt.title('daily receipts')
plt.show()
```

```python
from darts.dataprocessing.transformers import Scaler
scaler_series=Scaler()
series_scaled = scaler_series.fit_transform(series)

series_scaled.plot(label='daily receipts')
plt.legend();
```
